```python
import pandas as pd
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
# Apply average blur filter
# Adjust the kernel size to blur the image
ksize = (10, 10)
blurred_image = cv2.blur(img,ksize)

# Display the blurred image
display_image(img)
display_image(blurred_image)

# Save the blurred image
cv2.imwrite('blurred_image.jpg', blurred_image)
```

When choosing the kernel size for Gaussian blur, it's important to consider the size of the features we want to preserve in the image and the level of blurring we want to achieve. A larger kernel size will result in a stronger blur effect, which can potentially lead to loss of important details.

If we used a kernel size and found that it did not work as expected, it could be because the kernel size was too large for the features in the image.

Bilateral filtering is a non-linear filtering technique that reduces noise while preserving edges. It considers both the intensity differences and the spatial distances between pixels during the filtering process, resulting in noise reduction without significant blurring of sharp edges.


**d**	Diameter of each pixel neighborhood that is used during filtering. If it is non-positive, it is computed from sigmaSpace.


**sigmaColor**: Filter sigma in the color space. A larger value of the parameter means that farther colors within the pixel neighborhood (see sigmaSpace) will be mixed together, resulting in larger areas of semi-equal color.


**sigmaSpace**: Filter sigma in the coordinate space. A larger value of the parameter means that farther pixels will influence each other as long as their colors are close enough. When d>0, it specifies the neighborhood size regardless of sigmaSpace. Otherwise, d is proportional to sigmaSpace.

Part 3: Frequency Domain Filters

*Task 3.1: Low-Pass Filtering*

```python
# Read the input image in grayscale mode
img = cv.imread('/content/sample_data/flemingo.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "File could not be read, please check with os.path.exists()"

# Convert the input image to float32
img_float32 = np.float32(img)

# Perform the Discrete Fourier Transform (DFT)
dft = cv.dft(img_float32, flags=cv.DFT_COMPLEX_OUTPUT)

# Shift the zero frequency component (DC component) to the center
dft_shift = np.fft.fftshift(dft)

# Compute the magnitude spectrum (logarithmic scale)
magnitude_spectrum = 20 * np.log(cv.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

# Display the input image and its magnitude spectrum
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()


```

This code performs the Discrete Fourier Transform (DFT) on an input image and computes the magnitude spectrum.

1. Convert the input image to float32: It converts the input image img to the float32 data type using np.float32(). The DFT function in OpenCV expects the input image to be in the float32 format.
2. Perform the Discrete Fourier Transform (DFT): We apply the DFT to the img_float32 image using the cv.dft() function. The flag indicates that the output of the DFT should be complex numbers, consisting of real and imaginary parts.
3. Shifting the spectrum using np.fft.fftshift()  helps visualize the low-frequency components at the center and the high-frequency components at the edges.
4. Compute the magnitude spectrum (logarithmic scale): The magnitude spectrum represents the amplitude of the frequency components in the image. The magnitude is calculated using the cv.magnitude() function, which takes the real and imaginary parts of the shifted DFT as input. The resulting magnitude spectrum is then converted to a logarithmic scale using np.log() and multiplied by 20 to amplify the visualization.

The magnitude spectrum represents the amplitude of the frequency components in the image. In this code, the center of the spectrum appears brighter because it corresponds to the low-frequency components of the image. The rest of the spectrum appears gray because it represents the high-frequency components.

In an image, low-frequency components typically correspond to smooth and slowly varying regions, such as backgrounds or large objects. On the other hand, high-frequency components correspond to rapid changes in intensity, such as edges, textures, or fine details.

Bright spots in the spectrum indicate the presence of dominant frequency components or structures in the corresponding spatial region of the image. The gray regions represent areas with lower frequency content or less pronounced structures.

**Analyzing the magnitude spectrum can help in understanding the distribution and significance of different frequency components in an image and guide further image processing tasks, such as filtering, compression, or feature extraction.**

```python
# Read the input image in grayscale mode
img = cv.imread('/content/sample_data/flemingo.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "File could not be read, please check with os.path.exists()"

# Convert the input image to float32
img_float32 = np.float32(img)

# Perform the Discrete Fourier Transform (DFT)
dft = cv.dft(img_float32, flags=cv.DFT_COMPLEX_OUTPUT)

# Shift the zero frequency component (DC component) to the center
dft_shift = np.fft.fftshift(dft)

# Define the size of the Gaussian low-pass filter kernel
rows, cols = img.shape
crow, ccol = rows // 2, cols // 2
radius = 30  # Adjust this value to control the cutoff frequency of the low-pass filter

# Create a Gaussian low-pass filter kernel
mask = np.zeros((rows, cols, 2), np.uint8)
mask[crow - radius:crow + radius, ccol - radius:ccol + radius] = 1

# Apply the Gaussian low-pass filter kernel to the Fourier transformed image
dft_shift_filtered = dft_shift * mask

# Shift the zero frequency component back to the corner
dft_shift_filtered_back = np.fft.ifftshift(dft_shift_filtered)

# Convert the filtered image back to the spatial domain using the inverse Fourier Transform
img_back = cv.idft(dft_shift_filtered_back)
img_back = cv.magnitude(img_back[:, :, 0], img_back[:, :, 1])

# Normalize the result to the range [0, 255]
img_back = cv.normalize(img_back, None, 0, 255, cv.NORM_MINMAX)

# Display the original image and the low-pass filtered image
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_back, cmap='gray')
plt.title('Low-Pass Filtered Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

Low pass filter is the type of frequency domain filter that is used for smoothing the image. It attenuates the high-frequency components and preserves the low-frequency components.


This code performs the inverse Fourier Transform (IDFT) on the magnitude spectrum and restores the image.

1. The magnitude spectrum was previously converted to a logarithmic scale for visualization purposes. By taking the exponential of the magnitude spectrum divided by 20, we restore it to the original linear scale.
2. Shift the zero frequency component (DC component) back to the corner. This step is necessary before performing the inverse Fourier Transform.
3. Perform the inverse Fourier Transform (IDFT). It converts the frequency domain representation back to the spatial domain, resulting in the restored image.
4. It extracts the real and imaginary parts from the img_inverse image and calculates the magnitude using the cv.magnitude() function. The resulting img_restored image represents the amplitude of the restored image.
5. The img_restored image is normalized to the range [0, 255] using the cv.normalize() function. It ensures that the image is displayed correctly and can be saved as an 8-bit unsigned integer image (cv.CV_8U data type).



*Task 3.2: High-Pass Filtering*

```python
# Read the input image in grayscale mode
img = cv.imread('/content/sample_data/flemingo.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "File could not be read, please check with os.path.exists()"

# Convert the input image to float32
img_float32 = np.float32(img)

# Perform the Discrete Fourier Transform (DFT)
dft = cv.dft(img_float32, flags=cv.DFT_COMPLEX_OUTPUT)

# Shift the zero frequency component (DC component) to the center
dft_shift = np.fft.fftshift(dft)

# Define the size of the Gaussian high-pass filter kernel
rows, cols = img.shape
crow, ccol = rows // 2, cols // 2
radius = 30  # Adjust this value to control the cutoff frequency of the high-pass filter,
# smaller radius leads to a shaper image and bigger radius to a smoother image

# Create a Gaussian high-pass filter kernel (inverse of low-pass filter)
mask = np.ones((rows, cols, 2), np.uint8)
mask[crow - radius:crow + radius, ccol - radius:ccol + radius] = 0

# Apply the Gaussian high-pass filter kernel to the Fourier transformed image
dft_shift_filtered = dft_shift * mask

# Shift the zero frequency component back to the corner
dft_shift_filtered_back = np.fft.ifftshift(dft_shift_filtered)

# Convert the filtered image back to the spatial domain using the inverse Fourier Transform
img_back = cv.idft(dft_shift_filtered_back)
img_back = cv.magnitude(img_back[:, :, 0], img_back[:, :, 1])

# Normalize the result to the range [0, 255]
img_back = cv.normalize(img_back, None, 0, 255, cv.NORM_MINMAX)

# Display the original image and the high-pass filtered image
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_back, cmap='gray')
plt.title('High-Pass Filtered Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

This code applies a high-pass filter to the input image in grayscale mode using the Discrete Fourier Transform (DFT) and displays the result.
If we change the radius parameter, which determines the size of the central region of zeros in the high-pass filter kernel, it will affects the cutoff frequency of the high-pass filter.

- Reducing the radius value makes the central region of zeros smaller, leading to a higher cutoff frequency. This means that more high-frequency components will be preserved in the filtered image, resulting in a sharper image with more emphasized edges and details.

- Increasing the Radius: Increasing the radius value makes the central region of zeros larger, resulting in a lower cutoff frequency. As a result, fewer high-frequency components will pass through the filter, leading to a smoother image with reduced sharpness.

```python
# Apply Gaussian blur
gaussBlur = cv2.GaussianBlur(img, (5, 5), 0)
#Changing the size of the Gaussian blur kernel (increasing or decreasing the tuple (5, 5)) can lead to:
#Larger kernel sizes result in stronger blurring and may produce a smoother high-pass filtered image.
#Modifying the standard deviation parameter of the Gaussian blur operation affects the degree of blurring. Higher standard deviations result in more extensive blurring.

# Compute the high-pass filtered image
highPass = img - gaussBlur

# Optionally adjust the intensity of the high-pass filtered image
highPass = highPass + 70 * np.ones(img.shape, np.uint8)

# Display the original and high-pass filtered images side by side
display_image(np.hstack((img, highPass)))

```


This code performs high-pass filtering on the input image using a Gaussian blur operation.

- Gaussian Blur: The code first applies Gaussian blur to the input image using a kernel size of (5, 5) and a standard deviation of 0. This step is used to smooth the image and reduces noise, in order to serve as a low-pass filter.

- High-pass Filtering: The high-pass filtered image is obtained by subtracting the Gaussian-blurred image from the original input image. This operation emphasizes the high-frequency components in the image, such as edges and details, while attenuating low-frequency components.

- Intensity Adjustment: Optionally, the intensity of the high-pass filtered image can be adjusted. In this code, the intensity is increased by adding a constant value of 70 to every pixel in the filtered image. This adjustment enhance the contrast and visibility of high-frequency features in the image.

```python
# Read the input image
img = cv2.imread('/content/sample_data/flemingo.jpg', cv2.IMREAD_GRAYSCALE)

# Apply a Gaussian blur to create a smoothed version of the image
smoothed = cv2.GaussianBlur(img, (5, 5), 0)

# Subtract the smoothed image from the original to obtain the high-frequency components
high_pass = img - smoothed

# Add the high-frequency components back to the original image to sharpen it
sharpened = img + high_pass

# Clip the values to ensure they are within the valid range [0, 255]
sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)

# Display the original, smoothed, high-pass, and sharpened images
plt.subplot(221), plt.imshow(img, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(smoothed, cmap='gray')
plt.title('Smoothed Image'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(high_pass, cmap='gray')
plt.title('High-Pass (High-Frequency) Components'), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(sharpened, cmap='gray')
plt.title('Sharpened Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

High pass filter is the type of frequency domain filter that is used for sharpening the image. It attenuates the low-frequency components and preserves the high-frequency components.
Sharpening: The high-pass components are added back to the original image to enhance its sharpness. This process amplifies the contrast in edges and fine details, resulting in a sharpened appearance.

*Part 1: Linear Filters*

Task 1.1: Applying Average Blur:
   - **Effect**: Average blur is a type of low-pass filter that smooths the image by replacing each pixel's value with the average value of its neighboring pixels. It reduces high-frequency components in the image, resulting in blurring and smoothing.
   - **Effectiveness**: Average blur is effective for reducing noise. It can also be used to blur sensitive or private information in images.
   - **Utilization**: Average blur is used in photography or image editing applications to smooth out imperfections and reduce noise in images. For example, it can be applied to portrait photos to soften skin tones and reduce blemishes, resulting in a more flattering appearance.

   - **Effect**: Gaussian blur is also a low-pass filter that blurs the image.    - **Effectiveness**: Gaussian blur is effective for reducing image noise and detail while preserving the overall image structure. It is commonly used in image preprocessing tasks and can be adjusted by changing the standard deviation parameter to control the amount of blur.

   - **Effectiveness**: Sharpening filters are effective for enhancing image details and improving edge definition. They can be used to make images appear crisper and more visually appealing. However, caution should be exercised as excessive sharpening can lead to noise amplification and artifacts.
   - **Utilization**: Sharpening filters are used in image editing and enhancement tasks. For instance, in medical imaging, sharpening filters can be applied to X-ray or MRI images to enhance the visibility of fine structures and improve diagnostic accuracy.


*Part 2: Non-Linear Filters*

It is particularly effective at removing salt-and-pepper noise, where isolated pixels have extreme values that differ significantly from their neighbors.
   - **Effectiveness**: Median filtering is highly effective at removing impulse noise or random isolated outliers in an image. It preserves edges and fine details better than linear filters, making it useful for denoising images without blurring important structures.
   - **Utilization**: Median filtering is  useful for removing salt-and-pepper noise from images. It finds applications in various fields, including satellite imaging.

It applies a weighted average to the neighboring pixels based on their spatial and intensity differences. It can be considered as a combination of both low-pass and high-pass filtering.
   - **Effectiveness**: Bilateral filtering is effective at reducing noise while preserving sharp edges and fine details. It is useful in scenarios where noise reduction is desired without significant blurring of important image features.
   - **Utilization**: Bilateral filtering is commonly used in real-time video denoising applications. For example, in video conferencing systems, bilateral filtering can reduce noise while preserving the sharpness of facial features.
